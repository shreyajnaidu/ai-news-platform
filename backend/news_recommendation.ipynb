```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

import pandas as pd
```

```python
articles = [
    "OpenAI releases new GPT model for developers",
    "India wins cricket series against Australia",
    "AI startup raises $50 million in funding",
    "Global stock markets fall after inflation fears",
    "New machine learning breakthrough announced",
    "Celebrity buys luxury mansion in Los Angeles",
    "Tech companies invest heavily in artificial intelligence",
]

user_interest = (
    "artificial intelligence machine learning "
    "technology startups world news"
)
```

```python
all_text = [user_interest] + articles

vectorizer = TfidfVectorizer(
    stop_words="english"
)

tfidf_matrix = vectorizer.fit_transform(all_text)

print(tfidf_matrix.shape)
```

```python
user_vector = tfidf_matrix[0:1]

article_vectors = tfidf_matrix[1:]

scores = cosine_similarity(
    user_vector,
    article_vectors,
)[0]

scores
```

```python
results = pd.DataFrame({
    "Article": articles,
    "ML Score": scores,
})

results = results.sort_values(
    by="ML Score",
    ascending=False,
)

results
```

```python
results["ML Score"] = results["ML Score"].round(3)

results
```
